# Consolidating Data from API Call

```python
#Imports
import glob
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
FOLDER = "Data/"
file_list = sorted(os.listdir(FOLDER))
file_list
```

```python
## let's check the filepath 
FOLDER+file_list[5]
```

```python
#loading in a csv to ensure it works
pd.read_csv(FOLDER+file_list[5])
```

```python
import glob
## Make a filepath query
q = FOLDER+"final*.csv.gz"
print(q)
file_list = sorted(glob.glob(q))
file_list
```

```python
df = pd.concat([pd.read_csv(file) for file in file_list])
df.reset_index(inplace=True, drop=True)
df
```

```python
df = df[df['imdb_id'] != '0']
df
```

```python
df.to_csv("Data/final_tmdb_data_results_combined.csv.gz", compression="gzip", index=False)
```
